Классический начальный набор данных - данные пассажиров титаника.

```
PassangerId - уникальный номер пассажира

Name - имя человека, обращение к ней/нему
Sex - пол человека
Age - возраст на момент крушения
SibSp - количество братьев и сестер / супругов на борту "Титаника"

Cabin - каюта
```

````python
import pandas as pd
import scipy.stats as stats
````

````python
# Load dataset here
df = pd.read_csv("data.csv")
df.head(10)
````

1. Узнайте сколько было на борту человек

````python
cnt_passengers = df["PassengerId"].sum()
print(cnt_passengers)
````

2. Какой медианный и средний возраста пассажиров

````python
age_mean = df['Age'].mean()
print(age_mean)

age_median = df['Age'].median()
print(age_median)
````

3. Посчитайте процент выживаемости детей(до 16 лет) и взрослых

````python
children = df[(df['Age'] < 16)]
adults = df[(df['Age'] >= 16)]

children_survive_percent = children["Survived"].mean() * 100
adults_survive_percent = adults["Survived"].mean() * 100

print(f"{children_survive_percent} and {adults_survive_percent}")
````

4. Верна ли гипотеза, что женщины и дети сажались в шлюпки первыми и выживали больше?

````python
df['GroupOfPpl'] = df.apply(lambda x: 'Child' if x['Age'] <= 16 else x['Sex'], axis=1)

df_difference_surviving = df.groupby('GroupOfPpl')['Survived'].mean() * 100

contingency_table = pd.crosstab(df['GroupOfPpl'], df['Survived'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"{df_difference_surviving}")
print(p)
````

5. Зависит ли выживаемость от класса обслуживания?

````python
df['Pclass'].describe()

class_survival = df.groupby('Pclass')['Survived'].mean() * 100

contingency_table = pd.crosstab(df['Pclass'], df['Survived'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(p)
print(class_survival)
````

6. Посчитайте средний возраст умерших женщин и мужчин

````python
are_dead = df[df['Survived'] == 0]

avg_age_dead_men = are_dead[are_dead['Sex'] == 'male']['Age'].mean()

avg_age_dead_women = are_dead[are_dead['Sex'] == 'female']['Age'].mean()

print(f"men - {int(avg_age_dead_men)} and women - {int(avg_age_dead_women)}")
````

7. Различается ли процент выживаемости пассажиров с хотя бы одним родственником на борту и одиночек?

````python
df['HasRelatives'] = ((df['SibSp'] > 0) | (df['Parch'] > 0)).astype(int)

has_relatives_and_survived = df[df['HasRelatives'] == 1]['Survived'].mean() * 100

no_relatives_and_survived = df[df['HasRelatives'] == 1]['Survived'].mean() * 100

if (has_relatives_and_survived - no_relatives_and_survived) <= 0.1:
     print("No")
else:
    print('Yes')
````

8. Различается ли средняя стоимость билета у умерших и выживших пассажиров?

````python
fare_survived = df[df['Survived'] == 1]['Fare'].mean()
fare_dead = df[df['Survived'] == 0]['Fare'].mean() 

print(fare_survived - fare_dead)
````

9. Выведите максимальный и минимальный возраст выживших пассажиров во 2 классе обслуживания

````python
min_age_survived_secondClass = int(df[df['Pclass'] == 2]['Age'].min()) 
max_age_survived_secondClass = int(df[df['Pclass'] == 2]['Age'].max())

print(f"{min_age_survived_secondClass} + {max_age_survived_secondClass}")
````

10. Подтвердите или опровергните гипотезу: молодым мужчинам(от 18 лет) выжить легче, чем более взрослым

````python
male = df[df['Sex'] == 'male']

young_man = male[(male['Age'] >= 18) & (male['Age'] <= 35)]
older_man = male[male['Age'] > 35]

young_man_survival_rate = young_man['Survived'].mean() * 100
older_man_survival_rate = older_man['Survived'].mean() * 100

print(f"{young_man_survival_rate} + {older_man_survival_rate}")
````
